**Read in the texts of the State of the Union addresses, using the state_union corpus reader. Count occurrences of men, women and people in each document. What has hapened to the usages of these words over time?**

```python
from nltk.corpus import state_union 
cfd=nltk.ConditionalFreqDist( (target,fileid[:4]) 
                             for fileid in state_union.fileids() 
                             for w in state_union.words(fileid) 
                             for target in ['men','women','people'] if w.lower().startswith(target))

cfd.plot()
```

**Define a conditional frequency distribution over the Names Corpus that allowsyou to see which initial letters are more frequent for males versus females (see Figure 2-7).**


```python
from nltk.corpus import names
name_types = names.fileids()
male_names = names.words('male.txt')
female_names = names.words('female.txt')
```

What percentage of noun synsets have no hyponyms? You can get all noun syn-
  sets using wn.all_synsets('n').

```python
from nltk.corpus import wordnet as wn
nb_syn =  sum(1 for _ in wn.all_synsets('n'))
nb_syn
```

```python
fac = len([s for s in wn.all_synsets('n') if len(s.hyponyms()) == 0]) / nb_syn
fac
```
